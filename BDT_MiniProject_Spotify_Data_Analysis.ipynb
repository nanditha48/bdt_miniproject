```python
#COLUMN NAMES IN TRACKS.CSV
#popularity ranges 0-100
#duration in ms
#explicit column not bothered
#release date
#danceability: how is good is song for dancing based on a combination of musical elements like tempo, rythm, beat,strength etc , value ranges from 0-1
#energy: measure 0.0-1.0, perceptual measure of intensity and activity. higher the value , more energy
#key: pitch (notes of the song) total 11 keys(0-11)
#loudness: -60 to 0 db
#mode:major:1 , minor:0
#speechiness: presence of spoken words , podcasts, etc
#acousticness: 0-1 , 1 represents high confidence that the track is acoustic
#liveness: detects the presence of audience in recording
#valence: measure between 0-1, positiveness conveyed by a track
#data from 1922-2021
```

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv
df_tracks = pd.read_csv('/tracks.csv')
df_tracks.head()
```

```python
#null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df = df_tracks.sort_values('popularity', ascending="True").head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular =df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
df_tracks.duration.head()
```

```python
#first visualisation i.e correlational map (heatmap)
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")#setting the corr to pearson
plt.figure(figsize=(16,4))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlational HeatMap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#we welii 0.4% data of the total data
sample_df = df_tracks.sample(int(0.004 *len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#regression plot between loudness vs energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= "loudness", x="energy", color="c").set(title= "Loudness vs Energy Correlation")
#songs are in one direction, energy inc , loudness also inc
```

```python
#regression plot between popularity vs acousticness coorelation
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= "popularity", x="acousticness", color="b").set(title= "Popularity vs Acousticness Correlation")
#reg line is downwards ,popularity inc acousticness dec
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
#creating a new year column from the release date column
```

```python
#distribution plot to visualise total number in each year since 1922 that is available on spotify
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
#BarPlot showing year vs duration
total_dr = df_tracks.duration
fig_dims= (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation = 90)
```

```python
#reading spotify features csv
df_genre = pd.read_csv('/SpotifyFeatures.csv')
```
